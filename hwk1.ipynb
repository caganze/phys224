```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
%matplotlib inline
```

```python
import wisps
```

```python
#boltzmann constant 
k=1.3*(10**-19)*u.kilogram*(u.cm**2)/(u.s**2*u.Kelvin)
```

```python
def tau(rho, nh, T, a):
    m=rho*nh
    den=(nh*4*np.pi*a**2)
    num=np.sqrt(np.pi*m/(k*T))
    return num/den
```

```python
#now make the plots
rhos=np.array([10**-3, 1, 10**3])*(u.kilogram/u.cm**3)
nhs=np.array([10**-3, 1, 10**3])*(u.cm**-3)
ts=np.array([10, 100, 10**5])*(u.Kelvin)
```

```python
ass= np.logspace(-3, 1, 1000)*u.cm
```

```python
import seaborn as sns
sns.set_palette('rainbow')
```

```python
from itertools import combinations, permutations
```

```python
x, y, z=rhos, nhs, ts
```

```python
fig, ax=plt.subplots( figsize=(10, 6))


ax.plot(ass, tau(x[0], y[0], z[0], ass), label=r'$\rho$={},  $n_H$={}, T={}'.format(x[0], y[0], z[0]))
ax.plot(ass, tau(x[1], y[1], z[1], ass), label=r'$\rho$={},  $n_H$={}, T={}'.format(x[1], y[1], z[1]))
ax.plot(ass, tau(x[2], y[2], z[2], ass), label=r'$\rho$={},  $n_H$={}, T={}'.format(x[2], y[2], z[2]))

ax.plot(ass, tau(x[0], y[2], z[0], ass), label=r'$\rho$={},  $n_H$={}, T={}'.format(x[0], y[2], z[0]))
ax.plot(ass, tau(x[0], y[1], z[0], ass), label=r'$\rho$={},  $n_H$={}, T={}'.format(x[0], y[1], z[0]))

plt.yscale('log')


plt.tight_layout()
plt.minorticks_on()

plt.legend(fontsize=18)
plt.xlabel('a  (cm)', fontsize=18)
plt.ylabel(r'$\tau$ (s) ', fontsize=18 )
plt.savefig('figure_hwk1.pdf')
```
