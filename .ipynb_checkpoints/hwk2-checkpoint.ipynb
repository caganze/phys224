```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import wisps
from astropy.io import ascii
%matplotlib inline
```

```python
spec=ascii.read('orion_spectrum.dat').to_pandas()
spec.columns=['wave', 'flux']
```

```python
#normalize 
spec['flux']= spec.flux/np.nanmax(spec.flux)
```

```python
spec=spec.dropna()
```

```python
def fit_continuum():
    ##fit a continuum to the spec
    #do a sigma clip to reduce 
    flux=spec.flux.values
    sigma=np.nanstd(flux)
    mask=abs(flux) < 3*abs(sigma)
    print (mask)
    f=flux[mask]
    w=spec.wave.values[mask]
    plt.plot(w, f)
    pol=np.polyfit(w, f, 4)
    print (np.isnan(w).any(), np.isnan(f).any())
    return np.poly1d(pol)
```

```python
cont_f=fit_continuum()
```

```python
from tqdm import tqdm
```

```python
def measure_line_ratio(w1, w2, nsamples=1e4):
    vals=[]
    for i in tqdm(np.arange(int(nsamples))):
        vals.append(measure_ew(w1)/measure_ew(w2))
    return vals

```
